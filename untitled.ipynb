```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("salary_dataset.csv")
```

```python
#cмотрим на размер датафрейма
df.shape
```

```python
df.head(100)
```

```python
df.info()
```

```python
df["Age"] = df["Age"].astype("object")
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
#Заполняем медианами числовые значения
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Years of Experience'] = df['Years of Experience'].fillna(df['Years of Experience'].median())
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
df.info()
```

```python
#Так как количество пропусков не очень большое, заполняю, идентичными значениями из этого же столбца
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Education Level'] = df['Education Level'].fillna(df['Education Level'].mode()[0])
df['Job Title'] = df['Job Title'].fillna(df['Job Title'].mode()[0])
df.info()
df.isna().sum()
```

```python
#Смотрим как связанны между собой столбцы, видим, что данные Возраст, Зарплата и Стаж работы,
#тесно связанны между собой и имеют зависимость,в первую очередь рассмотрим взаимосвязь между ними

sns.pairplot(df[['Age', 'Gender', 'Education Level', 'Job Title', 'Years of Experience', 'Salary','Country','Race']])
```

Проведем двухфакторный анализ
Вопрос 1: Сначала посмотрим взаимосвязь между возрастом и зарплатой сотрудника?
Посмотрим выбросы

```python
#Так как взаимосвязь очевидна, могу прийти к выводу, что в возрасте от 20 до 30 уровень зарплаты редко привышает 150 тыс. рублей.
#Так же есть небольшие аномалии с совсем маленькой зарплатой в этом возрастном диапазоне, возможно это стажировка или небольшие проекты.
# От 20 до 40 лет, уровень ЗП напичает расти, возможно это связанно с опытом, возможно и с образованием.
# От 40 до 60 лет Уровень ЗП так же сохраняет стабильно небольшой рост, с небольшими аномалиями в стороны высокой ЗП,
#возможно зависит от должности, так же небольшой выброс наблюдается в возрасте от 50-60 с ЗП 150, возможно связан с отдельным проектом.
plt.figure(figsize=[8,5])
plt.scatter(data=df, x="Age", y="Salary")
plt.title("Взаимосвязь между возрастом и зарплатой")
plt.ylabel("Зарплата")
plt.xlabel("Возраст")
plt.show()
```

Вопрос 2 посмотрим взаимосвязь между возрастом и стажем работы? Посмотрим выбросы



```python
# Конечно очевидно, чем больше возраст, тем больше стаж работы, небольшие отклонения наблюдаются от 20-30 лет,
# возможно связанно с тем, что работать человек может начать еще со школы, и участвовать в проектах, отсюда и опыт.
# так же наблюдается меньше опыта, от 50-60 лет, возможно человек пришел в профессию позжно, лет в 40, например как я)
plt.figure(figsize=[8,5])
plt.scatter(data=df, x="Age", y="Years of Experience")
plt.title("Взаимосвязь между возрастом и стажем работы")
plt.ylabel("Стаж работы")
plt.xlabel("Возраст")
plt.show()
```

Вопрос 3: Посмотрим распределение между стажем и зарплатой есои взаимосвязь?



```python
#Взаимосвязь очевидна, чем больше опыта, тем выше зарплата как и в предыдущем случае.
plt.figure(figsize=[8,5])
plt.scatter(data=df, x="Years of Experience", y="Salary")
plt.title("Взаимосвязь между зарплатой и стажем работы")
plt.ylabel("Зарплата")
plt.xlabel("Стаж работы")
plt.show()
```

Посчитаем корреляцию Спирмена и построим матрицу.
Посмотрим взаимосвязь между возрастом, стажем и зарплатой.

```python
# ТАк же видна взаимосвязь высока между стажем и возрастом, а так же зависимоть зарплаты от опыта и возраста.
corr_matrix_spearman = df[["Age", "Salary", "Years of Experience"]].corr('spearman')
plt.figure(figsize=[8,5])
sns.heatmap(corr_matrix_spearman, annot=True, cmap='Reds')
plt.title("Матрица корреляции Спирмена")
plt.show()
```

Вопрос 4: Рассмотрим разницу зарплаты по половому признаку. Возможно ли что мужчины получают больше женщин?

```python
# Да- мужчины действительно получают больше зарплату, но не намного.
gender_df = df.groupby('Gender')[['Salary']].count()
gender_df
```

```python
gender_df.loc[['Female', 'Male', 'Other']].sort_values(by='Salary', ascending=False).plot(kind='bar',
                                                                  title='Зарплаты по гендерному признаку',
                                                                  xlabel='Зарплата',
                                                                  ylabel='Пол')
```

Вопрос5:Посмотрим статистику по странам, сразу посмотрю по половому признаку. Возможно в разных странах мужчины и женщины получают по разному зарплату и где-то женщины получают больше?
Построим график

```python
#Делаю выводы, что по всем представленным странам мужчины получают больше женщин, ориентировочно процентов на 30%
# Сортировка данных по убыванию
sorted_df = df.sort_values(by='Salary', ascending=False)

# Создание гистограммы с отсортированными данными
fig = px.histogram(
    sorted_df,
    x='Salary',
    y='Country',
    color='Gender',
    title='Уровень зарплат по странам и половому признаку'
)
fig.update_yaxes(categoryorder='total ascending')

fig.show()
```

Вопрос 6: Сделаем распределение зарплаты по 10 Топ-должностям и посмотрим выбросы

```python
#Сделаем боксплот
#Видно по графику, что медианная оценка больше всего у должностей Senior Project Engineer, а так же Software Engineer Manager, при это ЗП не самые большие
#Самая низкая медиана Software Engineer и  Back end Developer, причем уровень зарплат поход с предыдущими высокими медианами.
#Самая высокая зп достигает у Data Scientist Видно выбрала верное направление)))
#У должностей Software Engineer Manager и Маркетинкого менеджера, наблюдаются значительные выбросы, не могу сказать с ем связанны, возможно
#были провальные проекты.
top_job = df['Job Title'].value_counts().sort_values(ascending=False).head(10).index.values
top_job
sns.boxplot(x="Salary", y="Job Title",
            data=df[df['Job Title'].isin(top_job)])
plt.title('Распределение зарплаты по топ 10 должностям')
```
